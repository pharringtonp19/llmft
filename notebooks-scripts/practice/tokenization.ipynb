```python
text = "Based on the complaint, the key details are:\n\n1. Property Details:\n   - The property is located at 850 Hancock Avenue, Apt. 1B, Bridgeport, Connecticut.\n   - The property is leased to the defendant under a written yearly lease as part of the Federal Section 8 Housing Assistance Certificate Program.\n\n2. Tenant Details:\n   - The defendant is Anjelica Huertas, the tenant who has been leasing the property.\n\n3. Nature of the Complaint:\n   - The landlord, RH 850 HANCOCK AVE LP, alleges that the tenant's monthly portion of the rent ($254.00) has been due and unpaid for more than 10 days from March 1, 2022.\n   - The landlord gave the tenant proper notice to quit possession of the premises on or before April 13, 2022, but the tenant has continued to remain in possession and refuses to vacate the property.\n\n4. Unusual Details:\n   - The property is leased under the Federal Section 8 Housing Assistance Certificate Program, which is a government-subsidized housing program.\n   - The landlord has provided information about the tenant's right to counsel and available rental assistance programs, indicating that this is a standard part of the eviction process in Connecticut.\n\nIn summary, the landlord is seeking immediate possession of the premises and costs of the action due to the tenant's failure to pay their portion of the rent and their refusal to vacate the property after receiving proper notice."
```

```python
tokens = text.encode("utf-8")
tokens = list(map(int, tokens))
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]): 
        counts[pair] = counts.get(pair, 0) + 1 
    return counts 
```

```python
stats = get_stats(tokens)
print(stats)
```

```python
top_pair = max(stats, key=stats.get)
```

```python
def merge(ids, pair, idx): 
    newids = []
    i = 0 
    while i < len(ids): 
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]: 
            newids.append(idx)
            i += 2 
        else: 
            newids.append(ids[i])
            i += 1 
    return newids
```

```python
vocab_size = 276 
num_merges = vocab_size - 256 
ids = list(tokens)


merges = {}
for i in range(num_merges):
    stats = get_stats(ids)
    pair = max(stats, key=stats.get)
    idx = 256 + i 
    print(f"merging {pair} into a new token {idx}")
    ids = merge(ids, pair, idx)
    merges[pair] = idx 
```

```python
print("tokens length:", len(tokens))
print("ids length:", len(ids))
print("compression ratio", len(tokens)/len(ids))
```

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]

def decode(ids):
    tokens = b"".join(vocab[idx] for idx in ids)
    text = tokens.decode("utf-8", errors="replace")
    return text 
```
